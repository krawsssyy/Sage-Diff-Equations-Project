```python
t = var('t')
x = function('x')(t)
y = function('y')(t)
z = function('z')(t)
eq1 = diff(x, t) == x - y + z
eq2 = diff(y, t) == x + y - z
eq3 = diff(z, t) == (-1)*y + 2*z
syst = [eq1, eq2, eq3]
C1, C2, C3 = var('C1, C2, C3')
sol= desolve_system(syst, [x, y, z], [0, C1, C2, C3])
sol_x(t, C1, C2, C3) = sol[0].rhs()
sol_y(t, C1, C2, C3) = sol[1].rhs()
sol_z(t, C1, C2, C3) = sol[2].rhs()
soll_x = [sol_x(t, i, i - 1, i + 1) for i in range(1,10)]
soll_y = [sol_y(t, i, 3*i, 2*i) for i in range(1,10)]
soll_z = [sol_z(t, i, i, 1 - i) for i in range(1, 10)]
g1 = plot(soll_x, t, -10, 10, ymin=-20, ymax=20, color='red')
g2 = plot(soll_y, t, -10, 10, ymin=-20, ymax=20, color='blue')
g3 = plot(soll_z, t, -10, 10, ymin=-20, ymax=20, color='green')
show(g1 + g2 + g3)#1c
```
